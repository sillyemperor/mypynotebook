```python
%matplotlib inline

import dlib
import imageio
import matplotlib.pyplot as plt

image_file = './data/deeplab1.jpg'
img = imageio.imread(image_file)

# Locations of candidate objects will be saved into rects
rects = []
dlib.find_candidate_object_locations(img, rects, min_size=500)

fig = plt.figure()
ax = fig.add_subplot(111)
print("number of rectangles found {}".format(len(rects))) 
c = 0
for k, d in enumerate(rects):
#     print("Detection {}: Left: {} Top: {} Right: {} Bottom: {}".format(
#         k, d.left(), d.top(), d.right(), d.bottom()))
    if d.width() > 300 and d.height() > 300:
        rect = plt.Rectangle((d.left(), d.top()), d.width(), d.height(), fill=False)
        ax.add_patch(rect)
        c += 1
        if c > 3:
            break

plt.imshow(img)
plt.show()
    
```
